```python
df=pd.read_csv('/content/sample_data/california_housing_test.csv')
df.head()
```

```python
for i in range(3000):
  if df['housing_median_age'][i]<30:
    df['housing_median_age'][i]=0
  elif df['housing_median_age'][i]>30:
    df['housing_median_age'][i]=1

df
```

```python
for i in range(3000):## changing float type to str
  df['housing_median_age'][i]=str(df['housing_median_age'][i])

df.head()
```

```python
df['housing_median_age']=df['housing_median_age'].str.replace('[\.]','')

df.head()
```

```python
df['housing_median_age']=df['housing_median_age'].str[:1]

df.head()
```

```python

df['housing_median_age']= pd.to_numeric(df['housing_median_age'], errors='coerce')

type(df['housing_median_age'][0])
```

# Customization


**legend : showing representation of different points**

```python
house, room=plt.subplots(figsize=(10,9))
room.set(title='Graph')
df_scatter=room.scatter(
x=df['total_rooms'],
y=df['total_bedrooms'],
c=df['housing_median_age'])
room.legend(*df_scatter.legend_elements(),title='Housing_median_age');
```

**Plotting 2 oo plots**

```python
house, (room,room2)=plt.subplots(ncols=1,nrows=2,figsize=(10,9),sharex=True)## sharex is to share x axis to represent a the scale on
# lower axis as we are representing them in 1 col and 2 rows
room.set(title='Graph w.r.t total bedrooms',ylabel='y axis')
df_scatter=room.scatter(
x=df['total_rooms'],
y=df['total_bedrooms'],
c=df['housing_median_age'])
room.legend(*df_scatter.legend_elements(),title='Housing_median_age');

### Second plot
room2.set(title='Graph w.r.t latitude',xlabel=' x axis',ylabel='y axis')
df_scatter=room2.scatter(
x=df['total_rooms'],
y=df['latitude'],
c=df['housing_median_age'])
room.legend(*df_scatter.legend_elements(),title='latitude');

```

**Mean Line & changing color of graph & changing color of whole plot**

```python
house,room=plt.subplots(figsize=(10,9))
room.set(title='Graph w.r.t total bedrooms',ylabel='y axis', xlabel='x axis')
df_scatter=room.scatter(
x=df['total_rooms'],
y=df['total_bedrooms'],
c=df['housing_median_age'],
cmap='winter')# cmap is to change color according to scheme
room.legend(*df_scatter.legend_elements(),title='Housing_median_age')

## axhline is for meanline
room.axhline(y=df['total_bedrooms'].mean(),linestyle='--');## '-' is for solid line ##  mean line forms on y axis

##coloring fulll plot
plt.style.use('dark_background')## other colors : classic, seaborn-dark-palette

## naming the whole plot
house.suptitle('California Housing ',fontsize=32,fontweight='bold');
```
